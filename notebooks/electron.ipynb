### long TXF(short TMF) 時為上穿還是下穿？

```python
txf_realfill=cmd.book_printer('FUT_TAIFEX_TXF:202505',sdate=dt.date(2025,5,1),edate=dt.date(2025,5,8),depth=5,source='SRC_TAIFEX_RT_RD.ORIG_CPBQ.DEST_CPBQ')
```

```python
def detect_trade_through(fill_df, book_df):
    if book_df is None or book_df.empty:
        raise ValueError
    """
    判斷哪些成交穿價（成交價 > 對應時間的 ask_px）

    fill_df: DataFrame，包含 'time', 'px', 'sz'（成交時間、價格、數量）
    book_df: DataFrame，index 為 datetime，包含 'ask_px'
    回傳: fill_df 加上一欄 'through'（True/False）
    """
    fill_df = fill_df.copy()
    fill_df['time'] = pd.to_datetime(fill_df['time'])

    # 取得每筆成交時間點最近的 book.ask_px（前向填補）
    book_df = book_df[['ask_px']].copy()
    book_df = book_df[~book_df.index.duplicated()]
    book_df = book_df.sort_index()

    # reindex 成交時間對應的 ask_px
    ask_at_trade_time = book_df.reindex(fill_df['time'], method='ffill')
    fill_df['ask_px'] = ask_at_trade_time['ask_px'].values

    # 判斷穿價條件：成交價 > 該時刻的 ask_px
    fill_df['through'] = fill_df['px'] > fill_df['ask_px']
    return fill_df
```
